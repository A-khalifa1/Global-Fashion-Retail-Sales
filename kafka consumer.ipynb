```python
from kafka import KafkaConsumer
import json
import csv
import os

# Create a Kafka consumer instance
consumer = KafkaConsumer(
    'your_topic',  # The Kafka topic you're consuming from
    bootstrap_servers='localhost:9092',  # Ensure this is the correct address of your Kafka broker
    group_id='your_group_id',  # Consumer group ID
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))  # Deserialize the message from JSON
)

# Open the CSV file in append mode, even if the file is empty
file_exists = os.path.exists('saved_from_stream.csv')

# Open the CSV file to append the consumed data
with open('saved_from_stream.csv', mode='a', newline='', encoding='utf-8') as csv_file:
    fieldnames = [
        'Invoice ID', 'Line', 'Customer ID', 'Product ID', 'Size', 'Color', 'Unit Price', 'Quantity',
        'Date', 'Discount', 'Line Total', 'Store ID', 'Employee ID', 'Currency', 'Currency Symbol',
        'SKU', 'Transaction Type', 'Payment Method', 'Invoice Total'
    ]  # Add the fieldnames as per your data

    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write the header only if the file doesn't already exist
    if not file_exists:
        writer.writeheader()

    # Initialize a counter to track how many records have been consumed
    record_count = 0

    # Consume messages and append to the CSV file in real-time
    for message in consumer:
        if record_count >= 15:
            print("Reached 15 records. Stopping the consumer.")
            break  # Exit the loop after 15 records

        # Each message consumed is stored in message.value (a dictionary)
        data = message.value

        # Write the consumed data to the CSV file
        writer.writerow(data)
        csv_file.flush()  # Ensure the data is written immediately to the file
        print(f"Saved: {data}")  # Print to track progress

        # Increment the record count
```
